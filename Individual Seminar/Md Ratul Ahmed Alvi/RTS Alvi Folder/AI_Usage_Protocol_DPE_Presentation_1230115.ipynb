```python

import json
import re
import uuid
from datetime import datetime
from pathlib import Path

try:
    import ipywidgets as widgets
    from IPython.display import display, clear_output
except ImportError as e:
    raise ImportError("ipywidgets is required. Install with: pip install ipywidgets") from e

PROTOCOL_VERSION = "0.2"
verification_levels = [
    "none", "plausibility-check", "secondary-source-check",
    "primary-source-check", "methodological-check",
    "empirical-check", "multi-source-validation"
]
evaluation_statuses = ["accepted", "partially-accepted", "rejected", "revised", "unresolved"]
usefulness_levels = ["low", "medium", "high"]
usage_types = [
    "idea-generation", "literature-search-support", "summarization",
    "terminology-clarification", "comparison-structuring",
    "argument-critique", "text-revision", "translation",
    "code-generation", "verification-support", "modeling-support",
    "requirements engineering-support", "other"
]

prefilled_entries = json.loads('[{"objective": "Understand the assigned Dynamic Priority Exchange topic before preparing the seminar presentation.", "ai_tool": {"name": "ChatGPT", "model": "GPT-5.6 Thinking"}, "prompt": "Explain the Dynamic Priority Exchange server from the assigned Spuri and Buttazzo source in simple student-level language.", "ai_output_summary": "The AI clarified periodic and aperiodic tasks, EDF priorities, server period and capacity, deadline exchange, schedulability, and resource reclaiming.", "verification": {"actions": ["Compared the explanation with Sections 2 and 3 of the assigned Spuri and Buttazzo paper.", "Checked the definitions, task parameters, and equations against the primary source."], "level": "primary-source-check"}, "evaluation": {"status": "revised", "issues": ["Some initial explanations were simplified and required more precise terminology for capacity entries and deadline levels."], "usefulness": "high"}, "integration": {"used_in_work": true, "section": "Presentation background and DPE concept slides", "usage_type": "terminology-clarification", "direct_text_reused": false}, "reflection": "This interaction helped me understand the mechanism before deciding how to explain it orally. I revised simplified descriptions so that they matched the source."}, {"objective": "Select the source content that was relevant for a ten-minute seminar presentation.", "ai_tool": {"name": "ChatGPT", "model": "GPT-5.6 Thinking"}, "prompt": "Identify the essential pages and concepts from the source for a short presentation on the Dynamic Priority Exchange server.", "ai_output_summary": "The AI suggested focusing on the task model, the core exchange rule, the worked example, the utilization condition, resource reclaiming, implementation overhead, and one application example.", "verification": {"actions": ["Checked the recommended sections in the original paper.", "Compared the selected content with the professor\'s seminar requirements and presentation duration."], "level": "primary-source-check"}, "evaluation": {"status": "accepted", "issues": [], "usefulness": "high"}, "integration": {"used_in_work": true, "section": "Overall presentation scope", "usage_type": "summarization", "direct_text_reused": false}, "reflection": "The interaction helped reduce the topic to a manageable presentation scope without removing the formal and implementation aspects."}, {"objective": "Improve the logical order of the presentation.", "ai_tool": {"name": "ChatGPT", "model": "GPT-5.6 Thinking"}, "prompt": "Review my planned slide sequence and suggest a clearer flow for a ten-minute technical seminar presentation.", "ai_output_summary": "The AI recommended a problem-to-solution flow: motivation, task model, DPE mechanism, formal rules, scheduling example, schedulability, reclaiming, overhead, UPPAAL model, application, and conclusion.", "verification": {"actions": ["Compared the proposed sequence with the paper structure.", "Checked that the sequence covered the professor-required mathematical, algorithmic, implementation, and application elements.", "Adjusted the number of slides to fit the available presentation time."], "level": "methodological-check"}, "evaluation": {"status": "revised", "issues": ["The first suggested sequence contained too much detail for ten minutes and was shortened."], "usefulness": "high"}, "integration": {"used_in_work": true, "section": "Presentation structure", "usage_type": "comparison-structuring", "direct_text_reused": false}, "reflection": "I used the proposed sequence as planning support and manually reduced it to match the seminar time."}, {"objective": "Check the technical accuracy of the worked DPE example used in the presentation.", "ai_tool": {"name": "ChatGPT", "model": "GPT-5.6 Thinking"}, "prompt": "Check my explanation of the DPE worked example with task parameters tau1=(8,2), tau2=(12,3), server=(6,3), and the seven-unit aperiodic request at time 14.", "ai_output_summary": "The AI traced how capacity moves between deadline levels and how the seven-unit request is served using capacities of two, two, and three units.", "verification": {"actions": ["Checked all parameters and timing statements against the original example in the primary paper.", "Corrected the explanation of the capacity at deadline 12 so that it was described as one pool increasing over time."], "level": "primary-source-check"}, "evaluation": {"status": "revised", "issues": ["The first explanation skipped the idle interval and did not clearly show that the deadline-12 capacity was the same growing pool."], "usefulness": "high"}, "integration": {"used_in_work": true, "section": "Worked scheduling example", "usage_type": "verification-support", "direct_text_reused": false}, "reflection": "The interaction revealed an ambiguity in my explanation. I corrected the diagram and oral explanation after checking the original source."}, {"objective": "Improve grammar and simplify slide wording without changing the technical meaning.", "ai_tool": {"name": "ChatGPT", "model": "GPT-5.6 Thinking"}, "prompt": "Correct the grammar of my slide text and keep the wording short, clear, and suitable for a student seminar presentation.", "ai_output_summary": "The AI suggested shorter bullet wording, clearer headings, and removal of repeated or overly long sentences.", "verification": {"actions": ["Reviewed every language suggestion manually.", "Kept only wording that I understood and could explain during the presentation.", "Checked source-based technical statements after revision."], "level": "plausibility-check"}, "evaluation": {"status": "partially-accepted", "issues": ["Some suggestions sounded too formal for my speaking style and were simplified or rejected."], "usefulness": "high"}, "integration": {"used_in_work": true, "section": "Multiple presentation slides", "usage_type": "text-revision", "direct_text_reused": false}, "reflection": "The AI was used as a grammar and clarity reviewer. I retained control of the final wording and made sure I could explain every point."}, {"objective": "Improve diagram layout and visual readability.", "ai_tool": {"name": "ChatGPT", "model": "GPT-5.6 Thinking"}, "prompt": "Suggest a cleaner visual layout for the DPE capacity-flow diagram and identify overlapping, crowded, or unclear elements.", "ai_output_summary": "The AI proposed a horizontal process flow, increased spacing, shorter labels, controlled box widths, and separate visual treatment for stored capacities.", "verification": {"actions": ["Compared several generated layout suggestions.", "Rejected versions with overlaps, poor padding, or elements outside the slide area.", "Visually checked the final diagram at presentation scale."], "level": "empirical-check"}, "evaluation": {"status": "revised", "issues": ["Several early layout suggestions still caused overlap or overflow and were not used."], "usefulness": "high"}, "integration": {"used_in_work": true, "section": "DPE capacity-flow diagram", "usage_type": "requirements engineering-support", "direct_text_reused": false}, "reflection": "This was an iterative design process. I accepted only the layout ideas that remained readable and correctly represented the source example."}, {"objective": "Explain my own UPPAAL model accurately in the presentation.", "ai_tool": {"name": "ChatGPT", "model": "GPT-5.6 Thinking"}, "prompt": "Review screenshots of my UPPAAL model and help me explain the periodic-task, aperiodic-request, server automata, and verification queries.", "ai_output_summary": "The AI explained the purpose of each automaton and the meaning of the deadlock, completion, capacity, and deadline properties.", "verification": {"actions": ["Compared every explanation with my own model transitions and variables.", "Ran the verifier and confirmed the displayed properties passed.", "Corrected the scope description to state that the model focuses on resource reclaiming rather than implementing the complete DPE scheduler."], "level": "empirical-check"}, "evaluation": {"status": "revised", "issues": ["An initial interpretation overstated the model as a full DPE implementation."], "usefulness": "high"}, "integration": {"used_in_work": true, "section": "UPPAAL model and verification slides", "usage_type": "modeling-support", "direct_text_reused": false}, "reflection": "The interaction helped me state the model\'s contribution and limitation more accurately."}, {"objective": "Prepare for the seminar discussion and possible professor questions.", "ai_tool": {"name": "ChatGPT", "model": "GPT-5.6 Thinking"}, "prompt": "Act like a real-time systems professor and ask possible questions about DPE, the schedulability equation, implementation overhead, and my UPPAAL model.", "ai_output_summary": "The AI generated possible technical questions and challenged unclear explanations, especially about why exchanged capacity is safe, why it moves to a lower priority, and what the UPPAAL model does not represent.", "verification": {"actions": ["Prepared answers using the original paper and my own model.", "Rejected speculative questions or answers that were not supported by the source.", "Practiced explaining the utilization condition and implementation limitation in my own words."], "level": "methodological-check"}, "evaluation": {"status": "partially-accepted", "issues": ["Some generated questions went beyond the assigned scope and were not used."], "usefulness": "high"}, "integration": {"used_in_work": true, "section": "Presentation preparation and discussion", "usage_type": "argument-critique", "direct_text_reused": false}, "reflection": "This interaction helped me identify gaps in my oral explanation and prepare source-supported answers."}]')

metadata = {
    "protocol_version": PROTOCOL_VERSION,
    "student_id": "1230115",
    "course": "Real-Time Systems Seminar",
    "milestone": "Seminar presentation",
    "topic": "Dynamic Priority Exchange Server",
    "entries": []
}

student_id = widgets.Text(value=metadata["student_id"], description="Student ID:")
course = widgets.Text(value=metadata["course"], description="Course:")
milestone = widgets.Text(value=metadata["milestone"], description="Milestone:")
topic = widgets.Text(value=metadata["topic"], description="Topic:", layout=widgets.Layout(width="90%"))
metadata_box = widgets.VBox([
    widgets.HTML("<h3>Protocol Metadata</h3>"),
    student_id, course, milestone, topic
])
entries_box = widgets.VBox([])
status_output = widgets.Output()

def now_iso():
    return datetime.now().astimezone().isoformat(timespec="seconds")

def make_interaction_widget(data=None, parent_id=None):
    data = data or {}
    entry_id = "e-" + uuid.uuid4().hex[:8]

    def area(description, value="", height="90px"):
        return widgets.Textarea(
            description=description,
            value=value,
            layout=widgets.Layout(width="100%", height=height)
        )

    objective = area("Objective:", data.get("objective", ""), "70px")
    ai_tool_data = data.get("ai_tool", {})
    ai_tool = widgets.Text(value=ai_tool_data.get("name", "ChatGPT"), description="AI tool:")
    ai_model = widgets.Text(value=ai_tool_data.get("model", "GPT-5.6 Thinking"), description="Model:")
    prompt = area("Prompt:", data.get("prompt", ""), "115px")
    ai_output_summary = area("AI output summary:", data.get("ai_output_summary", ""), "100px")

    verification_data = data.get("verification", {})
    verification_level = widgets.RadioButtons(
        options=verification_levels,
        value=verification_data.get("level", "plausibility-check"),
        description="Verification level:"
    )
    verification_actions = area(
        "Verification actions:",
        "\n".join(verification_data.get("actions", [])),
        "105px"
    )

    evaluation_data = data.get("evaluation", {})
    evaluation_status = widgets.RadioButtons(
        options=evaluation_statuses,
        value=evaluation_data.get("status", "partially-accepted"),
        description="Status:"
    )
    usefulness = widgets.RadioButtons(
        options=usefulness_levels,
        value=evaluation_data.get("usefulness", "medium"),
        description="Usefulness:"
    )
    issues = area("Issues:", "\n".join(evaluation_data.get("issues", [])), "90px")

    integration_data = data.get("integration", {})
    used_in_work = widgets.Checkbox(
        description="Used in submitted work",
        value=integration_data.get("used_in_work", False)
    )
    section = widgets.Text(
        value=integration_data.get("section", ""),
        description="Section:",
        layout=widgets.Layout(width="85%")
    )
    usage_type = widgets.Dropdown(
        options=usage_types,
        value=integration_data.get("usage_type", "other"),
        description="Usage type:"
    )
    direct_text_reused = widgets.Checkbox(
        description="Direct text reused",
        value=integration_data.get("direct_text_reused", False)
    )
    reflection = area("Reflection:", data.get("reflection", ""), "105px")
    remove_button = widgets.Button(description="Remove interaction", button_style="danger")

    box = widgets.VBox([
        widgets.HTML("<hr><h3>Interaction " + entry_id + "</h3>"),
        objective,
        widgets.HBox([ai_tool, ai_model]),
        prompt,
        ai_output_summary,
        widgets.HTML("<b>Verification</b>"),
        verification_level,
        verification_actions,
        widgets.HTML("<b>Evaluation</b>"),
        evaluation_status,
        usefulness,
        issues,
        widgets.HTML("<b>Integration</b>"),
        used_in_work,
        section,
        usage_type,
        direct_text_reused,
        widgets.HTML("<b>Reflection</b>"),
        reflection,
        remove_button
    ])

    box.entry_id = entry_id
    box.parent_id = parent_id
    box.fields = {
        "objective": objective, "ai_tool": ai_tool, "ai_model": ai_model,
        "prompt": prompt, "ai_output_summary": ai_output_summary,
        "verification_level": verification_level,
        "verification_actions": verification_actions,
        "evaluation_status": evaluation_status, "usefulness": usefulness,
        "issues": issues, "used_in_work": used_in_work, "section": section,
        "usage_type": usage_type, "direct_text_reused": direct_text_reused,
        "reflection": reflection
    }

    def remove_entry(_):
        children = list(entries_box.children)
        if box in children:
            children.remove(box)
            entries_box.children = tuple(children)

    remove_button.on_click(remove_entry)
    return box

def add_interaction(_):
    entries_box.children = tuple(list(entries_box.children) + [make_interaction_widget()])

def serialize_entry(box):
    f = box.fields
    return {
        "id": box.entry_id,
        "timestamp": now_iso(),
        "parent_id": box.parent_id,
        "objective": f["objective"].value.strip(),
        "ai_tool": {"name": f["ai_tool"].value.strip(), "model": f["ai_model"].value.strip()},
        "prompt": f["prompt"].value.strip(),
        "ai_output_summary": f["ai_output_summary"].value.strip(),
        "verification": {
            "actions": [x.strip() for x in f["verification_actions"].value.splitlines() if x.strip()],
            "level": f["verification_level"].value
        },
        "evaluation": {
            "status": f["evaluation_status"].value,
            "issues": [x.strip() for x in f["issues"].value.splitlines() if x.strip()],
            "usefulness": f["usefulness"].value
        },
        "integration": {
            "used_in_work": bool(f["used_in_work"].value),
            "section": f["section"].value.strip(),
            "usage_type": f["usage_type"].value,
            "direct_text_reused": bool(f["direct_text_reused"].value)
        },
        "reflection": f["reflection"].value.strip()
    }

def safe_key(text):
    return re.sub(r"[^a-z0-9]+", "-", (text or "anonymous").lower()).strip("-")

def protocol_to_bibtex(protocol, json_filename):
    year = (protocol.get("exported_at") or now_iso())[:4]
    student = protocol.get("student_id") or "anonymous"
    milestone_name = protocol.get("milestone") or "Milestone"
    topic_name = protocol.get("topic") or "AI-assisted work"
    key = "ai-usage-protocol-" + safe_key(student) + "-" + year
    return "\n".join([
        "@misc{" + key + ",",
        "  author = {" + student + "},",
        "  title = {AI Usage Protocol for " + milestone_name + ": " + topic_name + "},",
        "  year = {" + year + "},",
        "  howpublished = {Submitted AI usage protocol notebook and JSON file},",
        "  note = {Documented reasoning episodes: " + str(len(protocol.get("entries", []))) +
        "; JSON file: " + json_filename + "}",
        "}"
    ])

def save_protocol(_):
    metadata["student_id"] = student_id.value.strip()
    metadata["course"] = course.value.strip()
    metadata["milestone"] = milestone.value.strip()
    metadata["topic"] = topic.value.strip()
    metadata["exported_at"] = now_iso()
    metadata["entries"] = [serialize_entry(box) for box in entries_box.children]

    base = "ai_usage_protocol_" + (metadata["student_id"] or "anonymous") + "_" + metadata["milestone"]
    base = base.replace(" ", "_").replace("/", "-")
    json_filename = base + ".json"
    bib_filename = base + ".bib"

    Path(json_filename).write_text(json.dumps(metadata, indent=2, ensure_ascii=False), encoding="utf-8")
    bibtex = protocol_to_bibtex(metadata, json_filename)
    Path(bib_filename).write_text(bibtex + "\n", encoding="utf-8")

    with status_output:
        clear_output()
        print("Saved protocol JSON:", json_filename)
        print("Saved BibTeX entry:", bib_filename)
        print("Entries:", len(metadata["entries"]))

add_button = widgets.Button(description="Add interaction", button_style="primary")
save_button = widgets.Button(description="Save protocol JSON", button_style="success")
add_button.on_click(add_interaction)
save_button.on_click(save_protocol)

entries_box.children = tuple(make_interaction_widget(item) for item in prefilled_entries)

display(metadata_box)
display(widgets.HBox([add_button, save_button]))
display(entries_box)
display(status_output)
```
