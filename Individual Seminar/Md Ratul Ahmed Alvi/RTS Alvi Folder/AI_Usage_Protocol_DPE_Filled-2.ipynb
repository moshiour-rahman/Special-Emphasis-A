```python

import json
import re
import uuid
from datetime import datetime
from pathlib import Path

try:
    import ipywidgets as widgets
    from IPython.display import display, clear_output
except ImportError as e:
    raise ImportError("ipywidgets is required. Install with: pip install ipywidgets") from e

PROTOCOL_VERSION = "0.2"

verification_levels = [
    "none", "plausibility-check", "secondary-source-check",
    "primary-source-check", "methodological-check",
    "empirical-check", "multi-source-validation"
]
evaluation_statuses = ["accepted", "partially-accepted", "rejected", "revised", "unresolved"]
usefulness_levels = ["low", "medium", "high"]
usage_types = [
    "idea-generation", "literature-search-support", "summarization",
    "terminology-clarification", "comparison-structuring",
    "argument-critique", "text-revision", "translation",
    "code-generation", "verification-support", "modeling-support",
    "requirements engineering-support", "other"
]

prefilled_entries = json.loads(r'''[{"objective": "Understand the Dynamic Priority Exchange server and the assigned research problem.", "ai_tool": "ChatGPT", "ai_model": "GPT-5.6 Thinking", "prompt": "Explain the Dynamic Priority Exchange server from the assigned Spuri and Buttazzo source in simple student-level language.", "ai_output_summary": "The AI explained periodic and aperiodic tasks, EDF, server period and capacity, and how unused capacity is preserved through deadline exchange.", "verification_level": "primary-source-check", "verification_actions": ["Checked the explanation against Sections 2 and 3 of the assigned Spuri and Buttazzo paper.", "Compared the terminology and equations with the original source."], "evaluation_status": "revised", "usefulness": "high", "issues": ["Some early explanations were simplified and needed more precise wording about capacity entries and deadline levels."], "used_in_work": true, "section": "Introduction and Background", "usage_type": "terminology-clarification", "direct_text_reused": false, "reflection": "The interaction helped me distinguish the DPE server from the complete real-time system and understand that capacity is transferred to a later deadline level."}, {"objective": "Identify the relevant parts of the assigned paper for a five-page seminar paper.", "ai_tool": "ChatGPT", "ai_model": "GPT-5.6 Thinking", "prompt": "Create a reading map showing which pages of the source are essential, optional, or not directly relevant for my Dynamic Priority Exchange seminar.", "ai_output_summary": "The AI highlighted the assumptions, DPE definition, schedulability proof, resource reclaiming, and implementation complexity as the essential sections.", "verification_level": "primary-source-check", "verification_actions": ["Opened the corresponding pages in the original PDF.", "Confirmed that the selected sections matched the professor's required topics."], "evaluation_status": "accepted", "usefulness": "high", "issues": [], "used_in_work": true, "section": "Overall research workflow", "usage_type": "summarization", "direct_text_reused": false, "reflection": "The reading map helped me focus on the scientific parts directly connected to my topic."}, {"objective": "Check and improve my own understanding notes.", "ai_tool": "ChatGPT", "ai_model": "GPT-5.6 Thinking", "prompt": "Review my self-written DPE notes, preserve my meaning, identify technical mistakes, and show what is still missing.", "ai_output_summary": "The AI confirmed correct parts and identified missing details in the formal exchange rule, the worked example, and the schedulability explanation.", "verification_level": "methodological-check", "verification_actions": ["Compared each correction with the assigned source.", "Kept my original examples where they were correct.", "Revised statements that mixed server utilization with individual aperiodic-task utilization."], "evaluation_status": "revised", "usefulness": "high", "issues": ["My first worked-example description was too general.", "My first schedulability explanation was not precise enough."], "used_in_work": true, "section": "DPE Algorithm, Schedulability, and Resource Reclaiming", "usage_type": "argument-critique", "direct_text_reused": false, "reflection": "I understood that the exchange rule requires simultaneous reduction of the active capacity and addition of the same amount at the periodic task's deadline."}, {"objective": "Describe and evaluate my own UPPAAL model.", "ai_tool": "ChatGPT", "ai_model": "GPT-5.6 Thinking", "prompt": "Analyze my UPPAAL screenshots and help me explain the periodic-task, aperiodic-request, and server automata and the verified properties.", "ai_output_summary": "The AI summarized the role of each automaton and explained the meaning of the deadlock, completion, capacity, and deadline properties.", "verification_level": "empirical-check", "verification_actions": ["Checked every described transition against my own UPPAAL model.", "Ran the verifier and confirmed that all listed queries passed.", "Confirmed that the model is a minimal resource-reclaiming abstraction rather than the complete DPE algorithm."], "evaluation_status": "revised", "usefulness": "high", "issues": ["The first interpretation was too close to a full DPE implementation and was corrected."], "used_in_work": true, "section": "Implementation and Evaluation", "usage_type": "modeling-support", "direct_text_reused": false, "reflection": "The interaction helped me explain the scope and limitation of my model honestly."}, {"objective": "Improve grammar, clarity, and scientific readability of my self-written content.", "ai_tool": "ChatGPT", "ai_model": "GPT-5.6 Thinking", "prompt": "Correct the grammar and improve the clarity of my own seminar-paper text while keeping my simple student-level tone and technical meaning.", "ai_output_summary": "The AI suggested grammar corrections, clearer transitions, more precise technical wording, and removal of repeated statements.", "verification_level": "plausibility-check", "verification_actions": ["Reviewed each suggested change manually.", "Rejected wording that sounded too advanced or changed my intended meaning.", "Checked that technical statements still matched the assigned source."], "evaluation_status": "partially-accepted", "usefulness": "high", "issues": ["Some suggestions were more formal than my natural writing style and were simplified."], "used_in_work": true, "section": "Multiple sections", "usage_type": "text-revision", "direct_text_reused": false, "reflection": "The AI was used as a language and clarity reviewer. I kept control over the final wording."}, {"objective": "Organize the paper according to IEEE seminar-paper conventions.", "ai_tool": "ChatGPT", "ai_model": "GPT-5.6 Thinking", "prompt": "Suggest an IEEE-style structure for a five-page seminar paper covering mathematical basics, the formal DPE algorithm, runtime overhead, UPPAAL implementation, and an application example.", "ai_output_summary": "The AI proposed a compact section order, page allocation, equation placement, figure captions, and IEEE-style citation locations.", "verification_level": "secondary-source-check", "verification_actions": ["Compared the structure with the professor's seminar requirements.", "Compiled the document with the IEEE conference template.", "Adjusted section lengths to remain within the five-page limit."], "evaluation_status": "revised", "usefulness": "high", "issues": ["The first outline had too many small subsections and was simplified."], "used_in_work": true, "section": "Overall paper structure", "usage_type": "comparison-structuring", "direct_text_reused": false, "reflection": "The interaction helped me create a coherent scientific flow and cover the professor-required elements."}, {"objective": "Improve the Overleaf layout and figure presentation.", "ai_tool": "ChatGPT", "ai_model": "GPT-5.6 Thinking", "prompt": "Fix LaTeX formatting problems such as overlapping TikZ elements, figure overflow, image placement, spacing, and unreadable merged screenshots.", "ai_output_summary": "The AI produced revised TikZ layouts, adjusted float placement, suggested width and spacing controls, and separated the four UPPAAL screenshots into individual figures.", "verification_level": "empirical-check", "verification_actions": ["Compiled each version in Overleaf.", "Checked all five pages visually for overlap, clipping, and unreadable labels.", "Repeated layout changes until no box or text extended outside the figure area."], "evaluation_status": "revised", "usefulness": "high", "issues": ["Several early TikZ layouts still overlapped or extended beyond the page and were rejected."], "used_in_work": true, "section": "Figures and document formatting", "usage_type": "code-generation", "direct_text_reused": false, "reflection": "This was an iterative validation process. The first generated layouts were not accepted."}, {"objective": "Evaluate external criticism and correct structural issues in the paper.", "ai_tool": "ChatGPT", "ai_model": "GPT-5.6 Thinking", "prompt": "Review criticism about duplicated sentences, orphaned figures, float placement, and gaps in the worked-example explanation, and apply only valid corrections.", "ai_output_summary": "The AI confirmed valid observations, removed a duplicate sentence, added figure and table references, improved transitions, and clarified capacity accumulation in the worked example.", "verification_level": "methodological-check", "verification_actions": ["Checked the worked example against the original Spuri and Buttazzo paper.", "Confirmed that each figure and table was mentioned in the text.", "Recompiled and visually reviewed the corrected paper."], "evaluation_status": "accepted", "usefulness": "high", "issues": [], "used_in_work": true, "section": "Worked Example, UPPAAL Evaluation, and Application Example", "usage_type": "argument-critique", "direct_text_reused": false, "reflection": "Only criticism supported by the source and the compiled document was applied."}]''')

metadata = {
    "protocol_version": PROTOCOL_VERSION,
    "student_id": "",
    "course": "Real-Time Systems Seminar",
    "milestone": "Seminar paper",
    "topic": "Dynamic Priority Exchange Server for Aperiodic Task Scheduling",
    "entries": []
}

student_id = widgets.Text(description="Student ID:", placeholder="Enter your ID")
course = widgets.Text(value=metadata["course"], description="Course:")
milestone = widgets.Text(value=metadata["milestone"], description="Milestone:")
topic = widgets.Text(value=metadata["topic"], description="Topic:", layout=widgets.Layout(width="90%"))

metadata_box = widgets.VBox([
    widgets.HTML("<h3>Protocol Metadata</h3>"),
    student_id, course, milestone, topic
])

entries_box = widgets.VBox([])
status_output = widgets.Output()

def now_iso():
    return datetime.now().astimezone().isoformat(timespec="seconds")

def make_interaction_widget(data=None, parent_id=None):
    data = data or {}
    entry_id = "e-" + uuid.uuid4().hex[:8]

    def ta(desc, value="", height="90px"):
        return widgets.Textarea(
            description=desc,
            value=value,
            layout=widgets.Layout(width="100%", height=height)
        )

    objective = ta("Objective:", data.get("objective", ""), "70px")
    ai_tool = widgets.Text(value=data.get("ai_tool", "ChatGPT"), description="AI tool:")
    ai_model = widgets.Text(value=data.get("ai_model", "GPT-5.6 Thinking"), description="Model:")
    prompt = ta("Prompt:", data.get("prompt", ""), "115px")
    ai_output_summary = ta("AI output summary:", data.get("ai_output_summary", ""), "100px")

    verification_level = widgets.RadioButtons(
        options=verification_levels,
        value=data.get("verification_level", "plausibility-check"),
        description="Verification level:"
    )
    verification_actions = ta(
        "Verification actions:",
        "\n".join(data.get("verification_actions", [])),
        "105px"
    )

    evaluation_status = widgets.RadioButtons(
        options=evaluation_statuses,
        value=data.get("evaluation_status", "partially-accepted"),
        description="Status:"
    )
    usefulness = widgets.RadioButtons(
        options=usefulness_levels,
        value=data.get("usefulness", "medium"),
        description="Usefulness:"
    )
    issues = ta("Issues:", "\n".join(data.get("issues", [])), "90px")

    used_in_work = widgets.Checkbox(
        description="Used in submitted work",
        value=data.get("used_in_work", False)
    )
    section = widgets.Text(
        value=data.get("section", ""),
        description="Section:",
        layout=widgets.Layout(width="85%")
    )
    usage_type = widgets.Dropdown(
        options=usage_types,
        value=data.get("usage_type", "other"),
        description="Usage type:"
    )
    direct_text_reused = widgets.Checkbox(
        description="Direct text reused",
        value=data.get("direct_text_reused", False)
    )
    reflection = ta("Reflection:", data.get("reflection", ""), "105px")
    remove_button = widgets.Button(description="Remove interaction", button_style="danger")

    box = widgets.VBox([
        widgets.HTML("<hr><h3>Interaction " + entry_id + "</h3>"),
        objective,
        widgets.HBox([ai_tool, ai_model]),
        prompt,
        ai_output_summary,
        widgets.HTML("<b>Verification</b>"),
        verification_level,
        verification_actions,
        widgets.HTML("<b>Evaluation</b>"),
        evaluation_status,
        usefulness,
        issues,
        widgets.HTML("<b>Integration</b>"),
        used_in_work,
        section,
        usage_type,
        direct_text_reused,
        widgets.HTML("<b>Reflection</b>"),
        reflection,
        remove_button
    ])

    box.entry_id = entry_id
    box.parent_id = parent_id
    box.fields = {
        "objective": objective,
        "ai_tool": ai_tool,
        "ai_model": ai_model,
        "prompt": prompt,
        "ai_output_summary": ai_output_summary,
        "verification_level": verification_level,
        "verification_actions": verification_actions,
        "evaluation_status": evaluation_status,
        "usefulness": usefulness,
        "issues": issues,
        "used_in_work": used_in_work,
        "section": section,
        "usage_type": usage_type,
        "direct_text_reused": direct_text_reused,
        "reflection": reflection
    }

    def remove_entry(_):
        children = list(entries_box.children)
        if box in children:
            children.remove(box)
            entries_box.children = tuple(children)

    remove_button.on_click(remove_entry)
    return box

def add_interaction(_):
    entries_box.children = tuple(list(entries_box.children) + [make_interaction_widget()])

def serialize_entry(box):
    f = box.fields
    return {
        "id": box.entry_id,
        "timestamp": now_iso(),
        "parent_id": box.parent_id,
        "objective": f["objective"].value.strip(),
        "ai_tool": {
            "name": f["ai_tool"].value.strip(),
            "model": f["ai_model"].value.strip()
        },
        "prompt": f["prompt"].value.strip(),
        "ai_output_summary": f["ai_output_summary"].value.strip(),
        "verification": {
            "actions": [x.strip() for x in f["verification_actions"].value.splitlines() if x.strip()],
            "level": f["verification_level"].value
        },
        "evaluation": {
            "status": f["evaluation_status"].value,
            "issues": [x.strip() for x in f["issues"].value.splitlines() if x.strip()],
            "usefulness": f["usefulness"].value
        },
        "integration": {
            "used_in_work": bool(f["used_in_work"].value),
            "section": f["section"].value.strip(),
            "usage_type": f["usage_type"].value,
            "direct_text_reused": bool(f["direct_text_reused"].value)
        },
        "reflection": f["reflection"].value.strip()
    }

def safe_bibtex_key(text):
    text = (text or "ai-usage-protocol").lower()
    return re.sub(r"[^a-z0-9]+", "-", text).strip("-") or "ai-usage-protocol"

def protocol_to_bibtex(protocol, json_filename=None):
    exported_at = protocol.get("exported_at") or now_iso()
    year = exported_at[:4]
    student = protocol.get("student_id") or "anonymous"
    milestone_name = protocol.get("milestone") or "Milestone"
    topic_name = protocol.get("topic") or "AI-assisted scientific work"
    json_file = json_filename or "ai_usage_protocol.json"
    entries_count = len(protocol.get("entries", []))
    key = "ai-usage-protocol-" + safe_bibtex_key(student) + "-" + year

    lines = [
        "@misc{" + key + ",",
        "  author = {" + student + "},",
        "  title = {AI Usage Protocol for " + milestone_name + ": " + topic_name + "},",
        "  year = {" + year + "},",
        "  howpublished = {Submitted AI usage protocol notebook and JSON file},",
        "  note = {Documented reasoning episodes: " + str(entries_count) + "; JSON file: " + json_file + "}",
        "}"
    ]
    return "\n".join(lines)

def save_protocol(_):
    metadata["student_id"] = student_id.value.strip()
    metadata["course"] = course.value.strip()
    metadata["milestone"] = milestone.value.strip()
    metadata["topic"] = topic.value.strip()
    metadata["exported_at"] = now_iso()
    metadata["entries"] = [serialize_entry(box) for box in entries_box.children]

    base_filename = "ai_usage_protocol_" + (metadata["student_id"] or "anonymous") + "_" + (metadata["milestone"] or "milestone")
    base_filename = base_filename.replace(" ", "_").replace("/", "-")
    json_filename = base_filename + ".json"
    bib_filename = base_filename + ".bib"

    Path(json_filename).write_text(
        json.dumps(metadata, indent=2, ensure_ascii=False),
        encoding="utf-8"
    )
    bibtex_entry = protocol_to_bibtex(metadata, json_filename=json_filename)
    Path(bib_filename).write_text(bibtex_entry + "\n", encoding="utf-8")

    with status_output:
        clear_output()
        print("Saved protocol JSON:", json_filename)
        print("Saved BibTeX entry:", bib_filename)
        print("Entries:", len(metadata["entries"]))

add_button = widgets.Button(description="Add interaction", button_style="primary")
save_button = widgets.Button(description="Save protocol JSON", button_style="success")
add_button.on_click(add_interaction)
save_button.on_click(save_protocol)

entries_box.children = tuple(make_interaction_widget(item) for item in prefilled_entries)

display(metadata_box)
display(widgets.HBox([add_button, save_button]))
display(entries_box)
display(status_output)
```

## Controlled Vocabulary

- `primary-source-check`: checked against the original research paper
- `methodological-check`: checked reasoning, assumptions, or formal correctness
- `empirical-check`: checked by running the model, compiler, verifier, or experiment
- `accepted`: used essentially as provided
- `partially-accepted`: used selectively
- `revised`: substantially corrected or rewritten
- `rejected`: discarded after evaluation
