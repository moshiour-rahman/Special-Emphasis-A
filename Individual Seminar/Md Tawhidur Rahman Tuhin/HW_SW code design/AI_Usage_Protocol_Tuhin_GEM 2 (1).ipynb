## Interaction 1 — Understanding ASAP, ALAP and the mobility window

**ID:** e-001
**AI tool:** Claude (Cowork mode)

### Objective
The GEM paper builds everything on two values per operation, ASAP and ALAP, and on the "mobility window" between them. I understood the definitions but wanted an intuitive grip on *why* mobility matters before I structured the presentation, because the whole geometric idea depends on it.

### Prompt I used
```
In high-level synthesis scheduling, what do ASAP and ALAP actually mean for an
operation, and what is the mobility window between them? Explain it simply, like
for a student, and tell me why mobility matters for scheduling.
```

### What AI gave me
Claude explained ASAP as the earliest cycle an operation can run (limited by the operations feeding it) and ALAP as the latest it can run without delaying the whole design. The gap between them is the mobility window — the operation's scheduling freedom. The key insight it gave: an operation with zero mobility is locked onto the critical path, while a wide window means the scheduler has room to move it to avoid resource clashes.

- Confirmed that zero-mobility operations form the critical path, as the paper states.
- **Verification level:** primary-source-check

### How I used it

- The "tight staircase = little freedom" framing I use verbally came from understanding mobility this way; the slide text is my own.
- **Used in slides:** Yes — conceptual foundation for the core-idea slide
- **Direct text reused:** No
- **Usage type:** terminology-clarification
- **Status:** accepted

### Reflection
Once I understood mobility as "scheduling freedom," the rest of GEM fell into place — point dominance, the staircase, and the conflict-resolution rule all reduce to how much freedom each operation has. That gave me the confidence to decide the order in which I present the concepts.

## Interaction 2 — Why scheduling is NP-hard only under resource limits

**ID:** e-002
**AI tool:** Claude (Cowork mode)

### Objective
The paper says scheduling is trivial with unlimited resources but NP-hard when resources are limited. I wanted to understand this distinction clearly so I could justify, on the "why is it hard" slide, why GEM is even necessary.

### Prompt I used
```
Why is operation scheduling in HLS easy with unlimited hardware but NP-hard when
resources are limited? Explain the difference simply and give a small concrete
example.
```

### What AI gave me
Claude explained that with unlimited units you just run every operation as soon as its inputs are ready (plain ASAP) — no conflict is possible. The hardness comes from sharing: with, say, two multipliers but six multiplications, you must decide which ones wait, and the number of valid assignments explodes combinatorially. It used the "two multipliers, six multiplications" example, which matches the HAL case in my paper.

### What I verified
- Confirmed against the seminar source that the resource-constrained version is the NP-hard one (referenced to De Micheli), not the unconstrained case.
- Checked that the example numbers (2 multipliers / 6 multiplications) are consistent with the HAL benchmark I use later.
- **Verification level:** primary-source-check

### How I used it
- Directly informed the "Why scheduling is hard" slide — the contrast between unlimited and limited resources, and the NP-hard label.
- The slide wording is mine; the AI only clarified the reasoning.
- **Used in slides:** Yes — the problem-motivation slide
- **Direct text reused:** No
- **Usage type:** terminology-clarification
- **Status:** accepted

### Reflection
This made me realise the talk needs to make the "limited resources" point explicitly, otherwise the audience won't see why a clever scheduler is needed at all. It shaped how I motivate the whole presentation.

## Interaction 3 — The core idea: mapping operations to points and non-crossing matching

**ID:** e-003
**AI tool:** Claude (Cowork mode)

### Objective
The central GEM idea is to turn each operation into a 2-D point (ASAP, ALAP) and solve scheduling as a non-crossing matching. I needed an intuitive explanation of *why* this geometric reformulation is valid, so I could present it as the heart of the talk and defend it in Q&A.

### Prompt I used
```
GEM plots each operation as a point (ASAP, ALAP) and then solves scheduling as a
"non-crossing matching" between operations and clock cycles. Explain in simple
terms why this works and what "non-crossing" means here.
```

### What AI gave me
Claude explained that operations on a dependency chain naturally line up as a rising staircase (each later operation has larger ASAP and ALAP), and that matching them to clock cycles in order — without "crossings" — is exactly the precedence constraint: an earlier operation must get an earlier-or-equal cycle. It noted the best such matching can be found efficiently with a sweep, in O(N log N).

### What I verified
- Checked the staircase and non-crossing-matching description against the GEM source.
- Confirmed the O(N log N) matching cost traces back to the geometric-matching reference (Aggarwal et al.), not invented by the AI.
- **Verification level:** primary-source-check

### How I used it
- This is the conceptual basis for my "core idea" and "two geometric ideas" slides.
- The staircase diagram and the slide text are my own; the AI clarified the underlying reasoning.
- **Used in slides:** Yes — core-idea and geometric-concepts slides
- **Direct text reused:** No
- **Usage type:** terminology-clarification
- **Status:** accepted

### Reflection
Understanding *why* non-crossing equals precedence let me present the geometry as a faithful reformulation rather than a gimmick. That distinction is exactly what a professor is likely to probe, so working it out beforehand was important.

## Interaction 4 — Clarifying the limits of GEM's optimality guarantee

**ID:** e-004
**AI tool:** Claude (Cowork mode)

### Objective
I needed to state precisely *where* GEM is optimal, because the advantages slide could easily overstate it. I wanted to be sure I could defend the exact claim in Q&A.

### Prompt I used
```
Is the GEM scheduling algorithm optimal? For the advantages slide I want to state
the optimality guarantee correctly.
```

### What AI gave me / where it was not enough
The first answer described GEM as "near-optimal and very fast" without clearly drawing the line between the per-chain guarantee and the full-graph result. That phrasing would have been misleading on a slide. I asked a follow-up specifically about chains versus the full DAG, and the corrected explanation was precise: the non-crossing matching is **provably optimal for a single dependency chain**, but once chains are merged in Phase 4 the greedy conflict resolution means the full-graph schedule can drift one or two cycles from the true optimum.

### What I verified
- Confirmed against the seminar source that the optimality guarantee is stated **per chain**, and that the full-DAG result is only "close to optimal."
- Checked that the Phase 4 merge is described as greedy with no global bound.
- **Verification level:** primary-source-check

### How I used it
- The advantages/limitations slide now says "optimal on chains" (not "optimal") and lists the full-graph drift as an explicit limitation.
- The wording is mine, written after the distinction was clarified.
- **Used in slides:** Yes — strengths & limitations slide
- **Direct text reused:** No
- **Usage type:** argument-critique
- **Status:** revised (first answer was imprecise; needed a follow-up)

### Reflection
This was the interaction where I most needed to be careful. A fluent but loose AI answer could have put an overstated claim on my slide. Forcing the chain-vs-DAG distinction and checking it against the source is what makes me able to defend that slide honestly.
