## Interaction 1 — Understanding EDS vs. EDL and why early idle time matters

**ID:** e-001
**AI tool:** Claude (Cowork mode)

### Objective
The whole EDL server rests on the difference between scheduling periodic tasks as soon as possible (EDS) and as late as possible (EDL). I understood the definitions from Section 6.5 of the Buttazzo book but wanted an intuitive grip on *why the position of idle time matters* before structuring the presentation.

### Prompt I used (representative — the original session's exact wording was not recorded)
```
In the EDL server from Buttazzo Section 6.5, what is the difference between EDS
and EDL, and why does it matter where the idle time is placed? Explain it simply,
like for a student.
```

### What AI gave me
Claude explained that both schedules meet exactly the same deadlines and leave the same total idle time, but EDS finishes the work early so its idle time collects at the end, while EDL delays every job to the latest safe moment so the idle time appears at the front. The key insight: an aperiodic request can only be served from its arrival onward, so only *early* idle time shortens its response — which is why the server is built on EDL and why the Chetto & Chetto theorem (EDL maximises idle time in every prefix [0, t]) is the foundation.

## Interaction 2 — Understanding the slack rule and the idle-time map

**ID:** e-002
**AI tool:** Claude (Cowork mode)

### Objective
The server decides *when* a request may run using the system slack and the precomputed idle map. I needed to understand the rule S(t) >= 1 and the meaning of the arrays E and Delta well enough to present the idle-map slide and answer questions about it.

### Prompt I used (representative)
```
How does the EDL server decide when an aperiodic request may run? Explain the
slack S(t) = min over d of [(d - t) - g(t, d)] and the idle-time arrays E and
Delta in simple terms.
```

### What AI gave me
Claude explained that with no request pending the periodic tasks simply run under EDF; when a request arrives, the server computes the EDL idle times of the current periodic instances. A time unit is usable exactly when the slack is at least one — each term of the minimum is a future deadline minus the periodic work that must be done by it. Because the idle pattern repeats every hyperperiod, it is stored as start times E and lengths Delta, and the server automatically gets the spare bandwidth Us = 1 - Up.

### What I verified

**ID:** e-004
**AI tool:** Claude (Cowork mode)

### Objective

### Prompt I used (representative)
```
For my guarantees and cost slides: state exactly what the EDL server guarantees.
Is it optimal in general? And what does the IPE server change?
```

### What AI gave me
Claude was precise about the scope: schedulability is exactly the plain EDF test (schedulable if and only if Up <= 1, Theorem 6.5), because requests use only would-be-idle time. The optimality (Lemma 6.3) says no *on-line* algorithm finishes a request sooner — it is a statement against on-line competitors, not clairvoyant ones. The exact server pays O(N n) to rebuild the idle map per arrival; the IPE server precomputes the map off-line and is *near*-optimal (it matches EDL in all but rare cases), not optimal.
